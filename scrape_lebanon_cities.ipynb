```python
from bs4 import BeautifulSoup
import requests
import re
import csv
import pandas as pd
```

```python
website_url = requests.get("https://ipfs.io/ipfs/QmXoypizjW3WknFiJnKLwHCnL72vedxjQkDDP1mXWo6uco/wiki/List_of_cities_and_towns_in_Lebanon.html").text
```

```python
soup = BeautifulSoup(website_url,'lxml')
tables = soup.findAll('table',{'class':'multicol'})
```

```python
cities = []
for table in tables:
    for li in table.findAll('li'):
        li_text = li.text.strip()
        li_text_cleaned = re.search('([^\&\-\(\,\/\:]+)',li_text).group()
        li_text_cleaned = re.sub(r"\s+$","",li_text_cleaned)
        cities.append(li_text_cleaned)
```

```python
cities
```

```python
cities.append("Beirut")
cities_df = pd.DataFrame(data = cities)
cities_df.to_csv("cities_lebanon.csv",index=False,header=False,sep=',')
```
